## Análise Temporal

```python
#conversão de dadas para anos
datas = df_ocorrencia_fator['ocorrencia_dia'].astype("datetime64")
df_ocorrencia_fator['ano']=datas.dt.year
datas = df_ocorrencia['ocorrencia_dia'].astype("datetime64")
df_ocorrencia['ano']=datas.dt.year
```

```python
#agrupando a base de dados usando o atributo 'ano' considerando soma e contagem
ocorrencias = df_ocorrencia_fator.groupby(['ano']).count().reset_index()
df_anos = df_ocorrencia_fator.groupby(['ano']).sum().reset_index()
```

```python
ocorrencias
```

```python
#inserindo atributos calculados
df_anos['percentual_operacional'] = df_anos['fator_operacional']/(df_anos['total_fatores'])
df_anos['ratio_operacional_humano'] = df_anos['fator_operacional']/(df_anos['fator_humano']+df_anos['fator_operacional'])
df_anos['total_ocorrencias_fator']=ocorrencias['codigo_ocorrencia']
df_anos['total_ocorrencias']=df_ocorrencia.groupby(['ano']).count().reset_index()['codigo_ocorrencia']
df_anos['ratio_ocorrencias']=df_anos['total_ocorrencias_fator']/df_anos['total_ocorrencias']
```

## Outros atributos

```python
def converter_hora(x):
    try:
        split = x.split(':')
        try:
            minutos = float(split[1])
            horas = float(split[0])
            if minutos > 30 and horas!=24:
                horas += 1
            elif minutos > 30 and horas==24:
                horas = 0
            return horas
        except Exception:
            return np.nan
    except Exception:
        return x
```

```python
df_ocorrencia['ocorrencia_hora']=df_ocorrencia['ocorrencia_hora'].apply(lambda x: converter_hora(x))
```

```python
sns.set(font_scale = 2.5)
sns_plot = sns.displot(data=df_ocorrencia, x="ocorrencia_hora", kde=True, height=10.5, aspect=1.)
sns_plot.set_axis_labels("Hora", "Ocorrências")
sns_plot.set(xticks=np.arange(0,25,2), yticks=np.arange(0,500,50))
sns_plot.savefig("hora.png")
```

```python
dia = df_ocorrencia[np.logical_not(df_ocorrencia['divulgacao_dia_publicacao'].isna())]['ocorrencia_dia']
dia_relatorio = df_ocorrencia[np.logical_not(df_ocorrencia['divulgacao_dia_publicacao'].isna())]['divulgacao_dia_publicacao']
```

```python
#considerando o primeiro dia como 01/01/2010
def dias_absoluto(x):
    split = x.split('/')
    return (int(split[0])-1)+(int(split[1])-1)*30+(int(split[2])-2010)*365

def dias_absoluto_2(x):
    split = x.split('-')
    return (int(split[2])-1)+(int(split[1])-1)*30+(int(split[0])-2010)*365
```

```python
dia = dia.apply(lambda x: dias_absoluto(x))
dia_relatorio = dia_relatorio.apply(lambda x: dias_absoluto_2(x))
```

```python
dia_relatorio
```

```python
tempo_relatorio = dia_relatorio - dia
```

```python
df_ocorrencia['tempo_relatorio'] = tempo_relatorio/30
```
